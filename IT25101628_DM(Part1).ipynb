```python
#Getting input from user Q3
x= float(input("Enter the First number : "))
y= float(input("Enter the Second number : "))

#getting input from user for opertaion
op = input("Enter operation :")
if op == "+":
    print("Result :",x+y)
elif op == "-":
    print("Result :",x-y)
elif op == "*":
    print("Result :",x*y)
elif op == "/":
    print("Result :",x/y)
else:
    print("Result: error")
        
```
