```python
# I. Compare two numbers Q1
x = 15
y = 10

# (a)
print("x > y :", x > y)

# (b)
print("x == y :", x == y)

# (c)
print("x != y :", x != y)

# II. Compare 10 and 20 using <=
print("10 <= 20 :", 10 <= 20)

# III. Check both conditions
print("5 > 3 and 10 > 5 :", 5 > 3 and 10 > 5)
```

```python
# BMI Calculator Q2

weight = float(input("Enter weight in kg: "))
height = float(input("Enter height in meters: "))

bmi = weight / (height * height)

if bmi < 18:
    category = "Underweight"
elif bmi <= 25:
    category = "Normal"
else:
    category = "Overweight"

print("BMI Value:", round(bmi, 2))
print("Classification:", category)

```

```python
# Student Grade Management System Q3

name = input("Enter student name: ")

mark1 = float(input("Enter marks for subject 1: "))
mark2 = float(input("Enter marks for subject 2: "))
mark3 = float(input("Enter marks for subject 3: "))

total = mark1 + mark2 + mark3
percentage = total / 3

if percentage >= 75:
    grade = "A"
elif percentage >= 60:
    grade = "B"
elif percentage >= 45:
    grade = "C"
elif percentage >= 30:
    grade = "D"
else:
    grade = "F"

print("\n----- Mark Sheet -----")
print("Student Name:", name)
print("Subject 1:", mark1)
print("Subject 2:", mark2)
print("Subject 3:", mark3)
print("Total Marks:", total)
print("Percentage:", round(percentage, 2), "%")
print("Grade:", grade)
```

```python
n = int(input("Enter a number: "))

sum = 0
for i in range(1, n + 1):
    sum += i

print("Sum of first", n, "natural numbers is:", sum)
```
